```python
# Importing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from sklearn.cluster import KMeans 
```

```python
# Inputs
x = 0.5
y = 0.5
z = 0.5
j = 'S'
```

```python
# Variables
N = 1000 # number of agents
t = 100 # cycles
ksi = 0.4 # percent of strategic agents in range from 0 to 50%
```

```python
# Agents

class Agent:
    def __init__(self, id, strategy, credibility):
        self.id = id
        self.strategy = strategy
        self.credibility = credibility
    def __str__(self):
        return 'Id: ' + str(self.id) + ', Strategy: ' + str(self.strategy) + ', Credibility: ' + str(self.credibility)  

####################################

agents = []

for index in range(int(N * ksi)):
    agents.append(Agent(index, 'S', 0))

agentsLength = len(agents)

for index in range(N - agentsLength):
    agents.append(Agent(agentsLength + index, 'H', 0))

np.random.shuffle(agents)
```

```python
# Choosing service suppliers
for cycle in range(t) :
    for agent in agents:
        setOfServiceSuppliersUniform = np.random.uniform(5, 15, 1)
        setOfServiceSuppliers = np.random.choice(agents, int(setOfServiceSuppliersUniform))

        print('##################\n')
        print('Cycle: ' + str(cycle) + ' | For agent: ' + str(agent) + '\n')

        print('Service suppliers: ')
        for serviceSupplier in setOfServiceSuppliers:
            print(serviceSupplier)

        p = 1
        r = 1

        if agent.strategy == 'S' and j == 'S':
            p = 1
            r = 1

        if agent.strategy == 'S' and j == 'H':
            p = y

        if agent.strategy == 'H' and j == 'H':
            p = 1 - x 
            r = 1 - x

        if agent.strategy == 'H' and j == 'S':
            r = z


        pt = np.minimum(agent.credibility, p)
        rt = np.minimum(pt, r)
        print('\nI: ' + str(agent.strategy) + ' | J: ' + str(j) + ' | R: ' + str(rt) + '\n')
```
