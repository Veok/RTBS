```python
# Importing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
import kmeans1d
import time
```

```python
# Inputs
x = 0.5
y = 0.5
z = 0.5
```

```python
# Variables
N = 1000 # number of agents
t = 100 # cycles
ksi = 0.4 # percent of strategic agents in range from 0 to 50%
```

```python
# Agents

class Agent:
    def __init__(self, id, strategy, credibility):
        self.id = id
        self.strategy = strategy
        self.credibility = credibility
    def __str__(self):
        return 'Id: ' + str(self.id) + ', Strategy: ' + str(self.strategy) + ', Credibility: ' + str(self.credibility)  

####################################

agents = []

for index in range(int(N * ksi)):
    agents.append(Agent(index, 'S', 1))

agentsLength = len(agents)

for index in range(N - agentsLength):
    agents.append(Agent(agentsLength + index, 'H', 1))

np.random.shuffle(agents)
```

```python
# Choosing service suppliers
for cycle in range(t) :
    print('Cycle: ' + str(cycle))
    start = time.time()
    for agent in agents:
        setOfServiceSuppliersUniform = np.random.uniform(5, 15, 1)
        setOfServiceSuppliers = np.random.choice(agents, int(setOfServiceSuppliersUniform))
           
        summarizedReputation = 0
        for serviceSupplier in setOfServiceSuppliers:
            p = 1
            r = 1
        
            if agent.strategy == 'S' and j == 'S':
                p = 1
                r = 1
        
            if agent.strategy == 'S' and j == 'H':
                p = y
        
            if agent.strategy == 'H' and j == 'H':
                p = 1 - x 
                r = 1 - x
        
            if agent.strategy == 'H' and j == 'S':
                r = z
        
            aij  = np.random.uniform(0, 1, 1)
            pt = np.minimum(aij, p)
            rt = np.minimum(pt, r)
            summarizedReputation += (serviceSupplier.credibility * rt)
        
        agent.credibility = summarizedReputation
        agentsCredibility = []
        
        for agentCreds in agents:
            agentsCredibility.append(agentCreds.credibility)
       
        agentsCredibility.sort()     
        clusters, centroids = kmeans1d.cluster(agentsCredibility, 2)

        agentIndexInCreds = agentsCredibility.index(agent.credibility)
        agentSetValue = clusters[agentIndexInCreds]
        indexesOfSimilar = []
        indexesOfOpposites = []
        similarValues = []
        oppositeValues = []
       
        for value in clusters:
            if value == agentSetValue:
                indexesOfSimilar.append(clusters.index(value))
            else:
                indexesOfOpposites.append(1)    

        for index in indexesOfSimilar:
            similarValues.append(agentsCredibility[index])


        for index in indexesOfOpposites:
            oppositeValues.append(agentsCredibility[index])

        agentsAvarage = sum(similarValues) / sum(oppositeValues)
        agent.credibility = agentsAvarage

    end = time.time()
    print('Elapsed time: ' + str(end - start))

for agent in agents:
    print(agent)

           
```
